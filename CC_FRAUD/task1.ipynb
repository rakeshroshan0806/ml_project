```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
test_df = pd.read_csv("/Users/rakesh/Desktop/python/fraudTest.csv")
train_df = pd.read_csv("/Users/rakesh/Desktop/python/fraudTrain.csv")
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
```

# 2. DROP UNNECESSARY COLUMNS

```python
drop_cols = ['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'merchant', 
             'first', 'last', 'street', 'city', 'state', 'job', 'dob', 'trans_num']

train_df.drop(columns=drop_cols, inplace=True)
test_df.drop(columns=drop_cols, inplace=True)
```


# 3. HANDLE CATEGORICAL DATA

```python
categorical_cols = train_df.select_dtypes(include=['object']).columns

le = LabelEncoder()
for col in categorical_cols:
    train_df[col] = le.fit_transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
X_train = train_df.drop('is_fraud', axis=1)
y_train = train_df['is_fraud']

X_test = test_df.drop('is_fraud', axis=1)
y_test = test_df['is_fraud']
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)

# Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
```

```python
try:
    print("\nTraining Random Forest...")
    rf = RandomForestClassifier(
        n_estimators=100, 
        random_state=42,
        n_jobs=-1,           # Use all CPU cores
        verbose=1,           # Show progress
        class_weight='balanced_subsample'  # Handle imbalance
    )
    
    rf.fit(X_train, y_train)
    print("Random Forest training complete")
    
    rf_pred = rf.predict(X_test)
    print("Random Forest predictions complete")
    
except Exception as e:
    print(f"Random Forest failed: {e}")
    # Fallback to a simpler model or skip
    rf_pred = None
```

```python
def evaluate_model(name, y_true, y_pred):
    print(f"\n{name} MODEL")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("Classification Report:\n", classification_report(y_true, y_pred))

evaluate_model("Logistic Regression", y_test, lr_pred)
evaluate_model("Decision Tree", y_test, dt_pred)
evaluate_model("Random Forest", y_test, rf_pred)
```

```python
results = test_df.copy()

results['Predicted_Fraud'] = rf_pred
results['Prediction_Label'] = results['Predicted_Fraud'].apply(
    lambda x: "FRAUD" if x == 1 else "NOT FRAUD"
)

print("\nSample Predictions:")
print(results[['is_fraud', 'Predicted_Fraud', 'Prediction_Label']].head(20))
```
